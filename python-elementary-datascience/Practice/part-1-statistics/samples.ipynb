### Запросы к данным в таблице

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[23, "M", 85, 60, 70], [18, "М", 45, 87, 25], [20, "Ж", 90, 70, 87]],
                  columns=["age", "gender", "math", "copm_science", "russian"], 
                      index=["Иванов И.И.", "Петров П.П.", "Сидорова С.С."])
new_one = pd.Series({"age":22, "gender":"Ж", "math":67, "chemistry":87, "russian":100}, name="Николаева Н.Н.")
df = df.append(new_one)
df
```

```python
# выборка данных с условием
print(df[df["russian"] > 50])
```

```python
# сложная выборка данных
print(df.query("((russian > 50) and (math > 50)) and (chemistry > 50 or copm_science > 50)"))
```

```python
# запрос с вычислением
print(df.query("russian + math > 100"))
```

```python
print(df.fillna(0.0))
```

```python
print(df.dropna())
```

```python
help(df.query)
```

### Дополнительно: работа со временем

```python
import datetime
dateA = datetime.date(year=2014, month=1, day=1)
dateB = datetime.date(2018, 5, 20)
dateC = datetime.date(1123, 12, 4)
print(dateA, dateB, dateC)
```

```python
# разница
print(dateB - dateA)
```

```python
# timedelta
td = dateB - dateA
print("Timedelta in days - %i or in seconds - %i" % (td.days, td.seconds))
```

```python
# время
timeA = datetime.time(12, 45, 22)
timeB = datetime.time(23, 11, 0, 9999)
print(timeA, timeB)
```

```python
print(timeB.utcoffset())
```

```python
# сегодня
print(datetime.date.today())
```
