## Построение спектрограммы, способ 2

```python
%matplotlib inline

# импорты
import numpy as np
import matplotlib.pyplot as plt
```

```python
# данные
## Число семплов
N = 600
## период сэмплов
T = 1.0 / 800.0
dt = 1.0 / N
phase = 0.0#np.pi/2.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x + phase) + 0.5*np.sin(80.0 * 2.0*np.pi*x + phase)
```

```python
# комплексные амплитуды и частоты
amp = np.fft.rfft(y)
freqs = np.fft.rfftfreq(x.shape[-1], T)
```

```python
# АЧХ
fig, ax = plt.subplots()
ax.grid()
ax.plot(freqs, 2.0/N * np.sqrt(amp.real**2+amp.imag**2))
plt.show()
```

```python
# ФЧХ
fig, ax = plt.subplots()
ax.grid()
ax.plot(freqs, 2.0/N * np.arctan2(amp.imag,amp.real))
plt.show()
```
