## Построение спектрограммы, способ 1

```python
%matplotlib inline

# импорты
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
```

```python
# данные
## Число семплов
N = 600
## период сэмплов
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
```

```python
# преобразование
yf = scipy.fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
```

```python
# АЧХ
fig, ax = plt.subplots()
ax.grid()
ax.plot(xf, 2.0/N * np.abs(yf))
plt.show()
```

```python
# ФЧХ
fig, ax = plt.subplots()
ax.grid()
ax.plot(xf, 2.0/N * np.arctan(yf[:N//2]))
plt.show()
```

```python
# ЛАЧХ
fig, ax = plt.subplots()
ax.grid()
ax.semilogy(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```
