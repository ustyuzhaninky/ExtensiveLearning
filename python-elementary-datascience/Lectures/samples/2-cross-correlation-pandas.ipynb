## Таблица кросс-корреляции в Pandas
Этот ноутбук показывает, как рассчитать таблицу кросс-корреляции, используя пакет pandas

```python
import pandas as pd
df = pd.read_csv("../../Practice/Datasets/XLMETH.csv", sep='\t').iloc[:,2:]
df = df.dropna()
df.head()
```

```python
# кросс-корреляция
df_corrs = df.corr()
df_corrs
```

```python
# Выбираем только те, которые больше заданного значения
df_selected = df_corrs[abs(df_corrs) > 0.5]
df_selected
```
